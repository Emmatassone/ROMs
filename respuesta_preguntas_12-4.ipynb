# ¿Cómo definir una familia de funciones?

```python
import numpy as np

#definir la función y sus parámetros

def f(x,mu,sigma=1): #Gaussianas
    return np.exp(-((x-mu)/sigma)**2)

#intervalo de interés para x
x=np.arange(-7,7,0.01)
#Crear una lista vacía
lista_de_funciones=[]
lista_de_parametros=np.arange(-1,1,0.2)
#iterar sobre el parámetro deseado (mu)
for mu in lista_de_parametros:
    lista_de_funciones.append(f(x, mu))

# Imprimimos la longitud de la lista de Gaussianas y la lista de parámetros. Ambas longitudes deberian coincidir.

if(len(lista_de_parametros)==len(lista_de_funciones)):
    print("Coinciden!")
```

# ¿Cómo graficamos la familia de funciones?

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
for mu in lista_de_parametros:
    plt.plot(x,f(x,mu), label='Gaussian con mu= '+str(round(mu,2)))
plt.legend()
plt.title("Familia de Gaussianas")
```

# ¿Cómo integrar funciones?

### **Opción a**: Tenemos la función a integrar **f** evaluada en una lista de puntos (Nuestro caso)

```python
#Calculamos las areas de cada gaussiana. Deberian ser iguales porque solo hicimos traslaciones

integral = np.trapz(lista_de_funciones, x)

# Imprimimos el resultado
print("El resultado de la integración es:", integral)
```

### **Opción b**: Tenemos la definición de **f** y no un conjunto de puntos

```python
from scipy.integrate import quad
limite_inferior = -7.
limite_superior = 7.
result, error = quad(f, limite_inferior, limite_superior, args=(lista_de_parametros[0]))

print("El resultado de la integración para el parámetro mu =",lista_de_parametros[0], "es:", result)
```

###### Notemos que hay una diferencia de área entre los dos métodos en los últimos decimales. Esto es porque cada método tiene una distinta precision. Al ser métodos de integración numérica, *no son exactos*
